```python
import requests
url = "https://stephen-king-api.onrender.com/api/books"
res = requests.get(url)
print(res)
```

```python
data = res.json()
data
```

```python
import pandas as pd
df = pd.json_normalize(data["data"])
df
```

```python
## Filtering Data
df = df[['id','Title','Year']]
df
```

```python
## Web Scrapping
import requests

url = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(url)
print(res)
```

```python
print(res.text)
```

```python
print(res.headers)
```

```python
if res.status_code == 200:
    print(res.content)
```

```python
with open("scraped_data/data1.html","w") as f:
    f.write(res.text) # scraped single page
```

```python
## BeautifulSoup(bs4)

from bs4 import BeautifulSoup
with open("scraped_data/data1.html","r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content,'lxml')
soup
```

```python
## BeautifulSoup Methods
soup.find("h1")
```

```python
soup.find("h3")
```

```python
all_h = soup.find_all("h3")

for h3 in all_h:
    print(h3.get_text(strip=True))
```

```python
## Usng try except
import requests
url = "https://stephen-king-api.onrender.com/api/shorts"
try:
    r = requests.get(url,timeout=3)
    r.raise_for_status()
except requests.exceptions.RequestException as e:
    print("Not Found..404!",e)
```

```python
## Headers (useful web scrapping)
url = "https://stephen-king-api.onrender.com/api/shorts"

headers = {
    "User-Agent":"Mozilla/4.0"
}
response = requests.get(url,headers=headers)
print(response.text)
```
