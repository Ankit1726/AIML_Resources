### Using beautifulsoup (BS4)

```python
!pip install lxml
```

```python
from bs4 import BeautifulSoup
with open("scraped_data/data1.html","r") as  f:
    html_content = f.read()

soup = BeautifulSoup(html_content,"html.parser")
```

```python
soup
```

```python
## bs4 methods
all_h3 = soup.find_all("h3")
all_countires = []

for h3 in all_h3:
    name = print(h3.get_text(strip=True))
    # print(h3.find_parent("div")['class'])

    population = h3.find_next("div").select("span.country-population")[0].get_text(strip=True)

    #print(h3.find_next("div").select_one("span.country-population").get_text(strip=True))
    
    all_countires.append([name,population])
```

```python
import pandas as pd
df = pd.DataFrame(all_countires,columns=["Name","Population"])
```

```python
df
```

```python
df.to_csv("cleaned_data/data.csv",index=False)
```
