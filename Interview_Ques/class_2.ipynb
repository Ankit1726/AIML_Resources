```python
##Ques_17 String Interpolation

name = 'Alice'
# using % formatting
print("Hello %s!" % name)

## usinf .format()
print("Hello {} Good Morning..!".format(name))

## using f string
print(f"Hello {name} Good Morning..!")
```

```python
## Ques_18 Singleton Pattern in Python
class Singleton:
     _instance = None

     def __new__(cls):
          if cls._instance is None:
               cls._instance = super().__new__(cls)
          return cls._instance
```

```python
## Swap two varible in Python
a,b = 5,10
print(f"{a} & {b}")
a,b = b,a
print(f"{a} & {b}")
```

```python
## Ques_18 Dunder ormagic methods in python
class Book:
     def __init__(self,pages):
          self.pages = pages

     def __add__(self,other):
          return Book(self.pages + other.pages)
     
book1 = Book(10)
book2 = Book(57)
book = book1+book2
print(book.pages)
```
