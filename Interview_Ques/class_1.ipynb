```python
 ## Write a function that takes a list of numbers and returns the sum.
def sum_lst(num):
     return sum(num)

print(sum_lst([1,2,3,4,5]))
```

```python
## Reverse a String
def reverse_str(str):
     return str[::-1]

print(reverse_str("Hello Good Morning"))
```

```python
## check Palindrome 
def is_palindrome(str):
     return str == str[::-1]

print(is_palindrome("nitin"))
print(is_palindrome("morning"))
```

```python
## remove duplicates from list
def remove_duplicates(lst):
     return list(set(lst))

print(remove_duplicates([1,2,1,2,4,5,7,8]))
```

```python
## Handling try and except
def divide(x,y):
     try:
          res= x/y
          return res
     except ZeroDivisionError:
          return "can not divide by zero..!"

print(divide(1,0))
```

```python
## lambda fucntion wit example
## lambda fucntion is small anonymous function it can take number of arguments but can only have one expression.

mul = lambda x,y: x*y

print(mul(4,4))
```

```python
## def fibonacci using recursion
def fibonacci(n):
     if n<=1:
          return n
     else:
          return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(10))
```

```python
## Decorators: It is a way to modify the function or extend the fucntionality of function without modifying the fucntion.

def my_decorator(func):
     def wrapper():
          print("calling first line..!")
          func()
          print("calling second line..!")
     return wrapper

@my_decorator
def say_hello():
     print("hello Good Morning..!")

say_hello()
```

```python
# *arg: allowsyou to pass variable number of positional arguments while **kwargs it allows to pass variable number of keyword arguments

def example(*args,**kwargs):
     for i in args:
          print(i,end=" ")
     for k,v in kwargs.items():
          print(f"{k}:{v}")

example(1,2,3,4,name="Alex",age=21)
```

```python
## Generator is an iterable that yields items one at a time using yield keyword. these are ore efficient for large dataset.ArithmeticError
def count(n):
     count = 1
     while count <= n:
          yield count
          count+=1

print(list(count(5)))
```

```python
## WAP to achieve the multithreading.
import threading
def print_num():
     for i in range(1,7):
          print(i,end=" ")

def print_letter():
     letters = 'Hello Good Morinng'
     for let in letters:
          print(let,end=" ")

t1 = threading.Thread(target=print_num)
t2 = threading.Thread(target=print_letter)

t1.start()
t2.start()

t1.join()
t2.join()
```
