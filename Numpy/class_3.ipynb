- Handling Missing values

```python
import numpy as np
```

```python
arr = np.array([1,2,np.nan,6,9,np.nan])
print(arr)
```

```python
print(np.isnan(arr))
```

- replace nan into num

```python
arr = np.array([1,2,np.nan,6,9,np.nan])
cleaned_arr = np.nan_to_num(arr,nan=10)
print(cleaned_arr)
```

- Handling infinite value

```python
arr = np.array([1,2,3,np.inf,4,-np.inf,7])
print(np.isinf(arr))
```

```python
cleaned_arr = np.nan_to_num(arr,posinf=100,neginf=-100)
print(cleaned_arr)
```
