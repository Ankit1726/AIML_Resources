- Python List vs NumPy Arrays

```python
import numpy as np
import time

## execution time comparision
size = 10_000_000  # 10M Dataset
Python_list = list(range(size))
st = time.time()
lst_sq = [x**2 for x in Python_list]
ed = time.time()
print(f"Python List Execution Time:{ed-st} seconds ")

## Numpy Arrays (much faster)
numpy_arr = np.array(Python_list)
st = time.time()
arr_sq = numpy_arr ** 2 # vector operation
ed = time.time()
print(f"Numpy Array Execution Time:{ed-st} seconds")
```

```python
## Memory Usage Comparison
import sys
print(f"Python List Size:{sys.getsizeof(Python_list)*len(Python_list)}")
print(f"Numpy Array Size:{numpy_arr.nbytes}")
```

- Operations on Numpy

```python
## Reshaping
arr = np.array([1,2,3,4,5,7])
print(arr.shape)

reshaped = arr.reshape((2,3))
print(reshaped,reshaped.shape)

## convert 2D -> 1D
flatten = reshaped.flatten()
print(flatten,flatten.shape)
```

```python
## Indexing
arr = np.array([1,2,3,4,5,7])
print(arr[2])
print(arr[0])
```

```python
## 2D Aray Indexing
arr = np.array([[1,2,3],[4,5,7],[2,8,9]])
print(arr[0][1])
print(arr[1][2])
```

```python
## Fancy Indexing
arr = np.array([1,2,3,4,5,7])
idx = [0,2,4]
print(arr[idx])
```

```python
## Boolean Indexing
print(arr[arr > 2])
print(arr[arr%2 == 0])
```

```python
## Slicing
arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[2:6]) 
print(arr[:6]) 
print(arr[3:])
print(arr[::2])
```

#### Copy v/s View  
- Slicing a list returns a copy but slicing a NumPy array returns a view(changes in original data).
   - Views are fast and memory-efficient (no data duplication). 
   - Copies are safe but slower and use more memory

```python
## sliced list is a copy
py_list = [1, 2, 3, 4, 5]
copy_list = py_list[1:4]
copy_list[1] = 333

print(copy_list)
print(py_list)
```

```python
## sliced numpy array is a view
np_arr = np.array([1, 2, 3, 4, 5])
view_arr = np_arr[1:4]
view_arr[1] = 333

print(view_arr)
print(np_arr)
```

```python
## Creating Copy
copy_arr = np_arr[1:5]
copy_arr[2] = 34
print(copy_arr)
print(np_arr)
```

- Operation Along Axis

```python
arr = np.array([[1,3,2],[4,5,7],[8,9,10]])
print(arr)
print(np.sum(arr)) # entire arr sum
```

```python
column_sum = np.sum(arr,axis=0)
print(column_sum) # sum along column
```

```python
row_sum = np.sum(arr,axis=1)
print(row_sum)
```

```python
## slicing
print(arr[0:3,1:3])
```

```python
## 3D Array Operation
arr = np.array([[[1,2,3],[4,5,7],[8,9,10]],[[1,2,3],[4,5,7],[8,9,10]]])

print(arr)
print(arr.shape)
```

```python
## Indexing
print(arr[0][1][1])   
print(arr[1][2][1]) 
```

```python
print(arr[:, :, 0]) # first col from both layers
print(arr[:, 0, :]) # first row from both layers
```

```python
## Manipulation Data
arr[:,0,:] = 12 # change first row
print(arr)
```

```python
arr[:,:,0] = 7 # change first column
print(arr)
```
