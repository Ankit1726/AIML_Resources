### NumPy Assignment Solution

```python
import numpy as np
```

```python
## Ques_1
arr = np.random.randint(1,21,size=(5,5))
print(f"Original Array: {arr}")

arr[:,2] = 1
print(f"Modified Array:{arr}")
```

```python
## Ques_2
arr = np.arange(1,17).reshape((4,4))
print(f"Original Array:{arr}")

np.fill_diagonal(arr,0)
print(f"Modified Array:{arr}")
```

```python
## Ques_3
arr = np.arange(1,37).reshape((6,6))
print(f"Original Array:{arr}")

sub_arr = arr[2:5,1:4]
print(f"SubArray:{sub_arr}")
```

```python
arr = np.random.randint(1,21,size=(5,5))
print(f"Original Array:{arr}")

border_el = np.concatenate((arr[0,:],arr[-1, :], arr[1:-1, 0], arr[1:-1, -1]))
print(f"Border Element:{border_el}")
```

```python
## Ques_5

array1 = np.random.randint(1, 11, size=(3, 4))
array2 = np.random.randint(1, 11, size=(3, 4))
print("Array 1:")
print(array1)
print("Array 2:")
print(array2)
addition = array1 + array2
subtraction = array1 - array2
multiplication = array1 * array2
division = array1 / array2

print("Element-wise addition:")
print(addition)
print("Element-wise subtraction:")
print(subtraction)
print("Element-wise multiplication:")
print(multiplication)
print("Element-wise division:")
print(division)
```
