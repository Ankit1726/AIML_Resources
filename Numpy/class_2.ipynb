### Common Data Types in Numpy
- Integers: int32, int64
- Floating point nums: float32, float64
- Boolean: bool
- Complex nums: complex64, complex128
- String: S (byte-str) & U (unicode-str)
- Object: generic python objects - object

```python
import numpy as np
```

```python
## Common Datatype
ar1 = np.array([1,2,3,4,5])
ar2 = np.array([1.0,2.0,5.0,7.0])
ar3 = np.array(["Python","DS","AIML"])

print(ar1.dtype)
print(ar2.dtype)
print(ar3.dtype)
print(ar1.nbytes)
```

```python
new_arr = ar1.astype("float64")
print(new_arr,new_arr.dtype)
```

```python
arr = np.array([1,2,3,4,5],dtype="float64")
print(arr,arr.dtype)
```

```python
##  Complex Numbers
arr1 = np.array([2 + 3j])
arr2 = np.array([5 + 8j])
print(arr1, arr1.dtype)
```

```python
print(arr1+arr2)
print(arr1-arr2)
```

```python
## Objects 
arr = np.array(["hello",{1,2,3},3.14])
print(arr,arr.dtype)
```

### Vectorization & Broadcasting

**Vectorization**: It is the process of applying operations to entire arrays or sequences of data at once, rather than iterating through individual elements using explicit loops. 

**Broadcasting**: It is a mechanism that allows NumPy to perform operations on arrays of different shapes without explicitly reshaping them (creating copies of the smaller array to match the larger array's shape). 

- Broadcasting Condition
   - For broadcasting to happen the dimensions should be compatible. 
   - Numpy compares shape element-wise. It starts dimension comparison with the trailing (i.e. rightmost) dimension and works its way left. 
   - Two dimensions are compatible when <ol> <li>they are equal, or</li> <li>one of them is 1. </li>

```python
## Vectorization
arr = np.array([1, 2, 3, 4, 5])

sq_arr = arr ** 2       
print(sq_arr)

arr2 = np.array([6, 7, 8, 9, 10])
print(arr + arr2)      

#Broadcasting
arr_mul10 = arr * 10  
print(arr_mul10)

arr1D = np.array([1, 2, 3])
arr2D = np.array([[1, 2, 3], [4, 5, 6]])
print(arr1D + arr2D)
```

```python
## Vector Normalization
arr = np.array([[1,2,3],[4,5,7]])
mean = np.mean(arr)
std_dv = np.std(arr)

print((arr-mean)/std_dv)
```

```python
## column wise
arr = np.array([[1, 2], [3, 4], [5, 6]])
mean = np.mean(arr, axis = 0)
std_dev = np.std(arr, axis = 0)
print((arr-mean) / std_dev)
```

### Mathematical Functions in NumPy
- Aggregation functions
   - sum()
   - prod()
   - min() & argmin() : min val & idx of min val
   - max() & argmax() : max val & idx of max val
   - mean()
   - median()
   - std() - standard deviation
   - var() - variance

```python
## Mathematical Functions
arr = np.array([1, 2, 3, 4, 5])

print(np.sum(arr))     
print(np.prod(arr))    
print(np.min(arr))    
print(np.argmin(arr))  
print(np.max(arr))     
print(np.argmax(arr))  
print(np.mean(arr))    
print(np.median(arr))  
print(np.std(arr))     
print(np.var(arr))   
```

- Power Functions
  - squre( )
  - sqrt( )
  - pow(a,b)

```python
arr = np.array([1,2,3,4,5,7])
print(np.square(arr))   
print(np.sqrt(arr))     
print(np.pow(arr, 3))
```

- Log & Exponential functions
   - log( ) - natual log
   - log10( ) - log base 10
   - log2( ) - log base 2
   - exp( ) - exponentional i.e. e^x

```python
print(np.log(arr))
print(np.log10(arr))
print(np.log2(arr))
print(np.exp(arr))
```

- Rounding functions
   - round( ) - rounds to nearest value
   - floor( ) - rounds down
   - ceil( ) - rounds up
   - trunc( ) - truncates decimal part

```python
print(np.round(2.678))    
print(np.floor(2.678))    
print(np.ceil(2.678))     
print(np.trunc(2.678))
```

- Additional functions
   - abs() - absolute value
   - sort() - sort all values (asc or desc)
   - unique() - returns unique values

```python
arr = np.array([1, 2, -5, 3, 8, -4, 2, 5])

print(np.abs(arr))     
print(np.sort(arr))  
print(np.unique(arr))  
```
