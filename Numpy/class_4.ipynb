- Indian Employee Dataset

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("indian_employee_data.csv")
df.head()
```

```python
## Missing value
df.isnull().sum()
```

```python
df['Experience_Years'].fillna(df['Experience_Years'].mean(),inplace=True)
df['Performance_Rating'].fillna(df['Performance_Rating'].mean(),inplace=True)
df['Salary_INR'].fillna(df['Salary_INR'].mean(),inplace=True)
```

```python
df.replace([np.inf,-np.inf],np.nan,inplace=True)
```

```python
## remove duplicates
df.drop_duplicates(inplace=True)
```

```python
## replace negative salary
df['Salary_INR'] = np.where(df['Salary_INR']<0,df['Salary_INR'].mean(),df['Salary_INR'])
```

```python
sal_mean = df['Salary_INR'].mean()
sal_std = df['Salary_INR'].std()
lower = sal_mean - (3*sal_std)
upper = sal_mean + (3*sal_std)
```

```python
## remove rows where salary to high or to low
df = df[(df['Salary_INR'] >=lower) & (df['Salary_INR']<=upper)]
df.to_csv("cleaned_data.csv",index=False)
print("Data Cleaning Completed..!")
```
