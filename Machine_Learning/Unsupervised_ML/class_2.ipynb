```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
iris
```

```python
import pandas as pd
## extracting dataset
X = pd.DataFrame(iris.data)
X.columns = iris.feature_names
```

```python
X = iris.data
y = iris.target
```

```python
## Visualize
sns.scatterplot(x=X[:,0],y=X[:,2],c=y)
```

```python
## Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
x_scaled
```

```python
## Optinal: dimensionality reduction using PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(x_scaled)
```

```python
## Elbow Method
wcss = []

for k in range(1,11):
     kmeans = KMeans(n_clusters=k)
     kmeans.fit(pca_data)
     wcss.append(kmeans.inertia_)
```

```python
print(kmeans.inertia_)
```

```python
## Kneed Module (automatically find value of K for making clusters)
from kneed import KneeLocator
knee = KneeLocator(range(1,11),wcss,curve="convex",direction="decreasing")
print(f"Optimal K={knee.elbow}")
```

```python
sns.lineplot(x=range(1,11),y=wcss,marker='o')
```

```python
## KMeans
kmeans = KMeans(n_clusters=3,random_state=42)
labels = kmeans.fit_predict(pca_data)
sns.scatterplot(x=pca_data[:,0],y=pca_data[:,1],c=labels)
sns.scatterplot(x=kmeans.cluster_centers_[:,0],y=kmeans.cluster_centers_[:,1],marker="x",c="red",s=300)
```
