- Unsupervised ML
   - DBSCAN Clustering
      - It stands for Density Based Spatial Clustering of Application with Noise
      - It also used for Anamoly Detection

```python
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()

X = iris.data
y = iris.target
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
## Visualize
sns.scatterplot(x=x_scaled[:,0],y=x_scaled[:,2],c=y)
```

```python
## DBSCAN
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(
     eps=0.8,
     min_samples=5
)
labels = dbscan.fit_predict(x_scaled)
```

```python
sns.scatterplot(x=x_scaled[:,0],y=x_scaled[:,2],c=y)
```

- Non Linear Data

```python
from sklearn.datasets import make_moons

X,y = make_moons(
     n_samples=400,
     noise=0.05,
     random_state=42
)
```

```python
## Scaling
x_scaled = scaler.fit_transform(X)

sns.scatterplot(x=x_scaled[:,0],y=x_scaled[:,1])
```

```python
## Kmeans
from sklearn.cluster import KMeans

kmeans = KMeans(
     n_clusters=2,
     random_state=42
)
labels = kmeans.fit_predict(x_scaled)
```

```python
sns.scatterplot(x=x_scaled[:,0],y=x_scaled[:,1],c=labels)
```

```python
## DBSCAN
dbscan = DBSCAN(
     eps=0.5,
     min_samples=5
)
labels = dbscan.fit_predict(x_scaled)
```

```python
sns.scatterplot(x=x_scaled[:,0],y=x_scaled[:,1],c=labels)
```
