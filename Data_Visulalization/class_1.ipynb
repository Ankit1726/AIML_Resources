```python
## Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] 
```

```python
plt.plot(oscar_years,oscar_revenue)
plt.title('Oscar Years V/S Revenue(in M$)')
plt.xlabel('Years')
plt.ylabel('Revenue (in M$)')
```

```python
## Plot Multiple
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]
```

```python
plt.plot(oscar_years, oscar_revenue, label="Oscar Winners")
plt.plot(non_oscar_years, non_oscar_revenue, label="Non-Oscar Winners")
plt.title("Years vs Revenue(in $M)")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")
plt.legend()
```

```python
# Bar charts
movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232] 
```

```python
## Bar Charts
plt.bar(years,revenue,color='purple')
plt.xlabel("Years")
plt.ylabel("Revenue (in M$)")
plt.title('Yearly Revenue Of Movies')

## Add Label on top of bar
for i in range(len(revenue)):
    # text(pos of text acc to x-axis, pos of y, text, alignment)
    plt.text(years[i],revenue[i]+20,str(revenue[i]),ha='center')

plt.ylim(0,max(revenue)+200)
# (lower,upper limit - add space)
```

```python
### comparing data with bar charts
oscar_movies = ["The Dark Knight","The Hurt Locker","The King's Speech","The Artist","Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]
```

```python
## Ploting Bar Chart
import numpy as np
x = np.arange(len(years))
width = 0.4

plt.bar(x-width/2,oscar_revenue,width,label='Oscar',color='green')
plt.bar(x+width/2,non_oscar_revenue,width,label='Non-Oscar',color = "red")

plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.title("Yearly revenue of Oscar vs Non-Oscar movies")
plt.legend()
plt.tight_layout()
plt.xticks(x, years) 
# To show x-axis labels
```

```python
# Scatter Plots
people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]
```

```python
plt.scatter(age,blood_pressure,color='green')
plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
```

```python
## Customizing Plots
colors = ['green' if bp<135 else "red" for bp in blood_pressure]
sizes = blood_pressure
plt.scatter(age, blood_pressure, color=colors, s=sizes, marker='^', alpha=0.5)
plt.scatter(age, blood_pressure, c=blood_pressure, cmap="OrRd")
## makes more sense
plt.colorbar(label="BP")
plt.grid(True)
for i in range(len(age)):
    plt.annotate(people[i], xy=(age[i], blood_pressure[i]))
plt.xlim(min(age), max(age)+10)
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
```

```python
plt.pie(amounts,labels=expenses)

## set Style
plt.style.use("ggplot")

# add %-age of each slice
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%')

#set colors
colors = ["purple", "green", "blue", "pink", "yellow"]
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%', colors=colors)

# wedgeprops
plt.pie(amounts, labels=expenses,  autopct='%1.1f%%', 
        wedgeprops={"edgecolor":"black", "linewidth": 1, "linestyle":"-.", "fill":False}
       )
```

```python
# explode
explode = [0, 0, 0, 0.2, 0]
# angle
plt.pie(amounts, labels=expenses, explode=explode, shadow=True, startangle=90) # top
plt.pie(amounts, labels=expenses, explode=explode, shadow=True, startangle=180) # flips
```
