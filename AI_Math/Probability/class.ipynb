```python
## Binomial Distribution
from scipy.stats import binom
n = 5
k = 3
p = 0.5

prob = binom.pmf(k, n, p)
print(prob)
```

```python
## Uniform Distribution
import numpy as np
import matplotlib.pyplot as plt

## values from 0 to 10
values = np.random.uniform(0,10,10_000)

## Ploting Histogram
plt.hist(values,bins=100,density=True,alpha=0.4,color="green")
plt.title("Continuous Uniform Distribution between 0 and 10")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)

plt.show()
```

```python
## Normal Distribution

import numpy as np

import matplotlib.pyplot as plt
from scipy.stats import norm

## parameters
mu = 70 # mean
sigma = 10  # std_dev

## x-axis values
x = np.linspace(30,100,700)

## Normal PDF
y = norm.pdf(x,mu,sigma)

## plot bell curve
plt.plot(x,y,color='black',linewidth=1.5,label='Normal Distribution')

## Shade 68% region (μ ± 1σ)
x1 = np.linspace(mu - sigma, mu + sigma, 1000)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color='green', alpha=0.3, label='68% region (1σ)')

# Shade 95% region (μ ± 2σ)
x2 = np.linspace(mu - 2*sigma, mu + 2*sigma, 1000)
plt.fill_between(x2, norm.pdf(x2, mu, sigma), color='yellow', alpha=0.2, label='95% region (2σ)')

# Labels and title
plt.title("Normal Distribution of Student Scores")
plt.xlabel("Score")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

# Shade 95% region (μ ± 2σ)
x2 = np.linspace(mu - 2*sigma, mu + 2*sigma, 1000)
plt.fill_between(x2, norm.pdf(x2, mu, sigma), color='yellow', alpha=0.2, label='95% region (2σ)')
```
