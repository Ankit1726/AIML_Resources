```python
## plot Graph for  y = √x
import numpy as np
import matplotlib.pyplot as plt

## Define x values (x>=0 for sqrt)
x = np.linspace(0,100,100_000)
y = np.sqrt(x)

## Plot
plt.plot(x, y, label="y = √x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = √x")
plt.legend()
plt.grid(True)
plt.show()
```

```python
## Create x values from -2π to 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)

## Plot the graph
plt.plot(x, y, label='y = sin(x)', color='blue')

## Highlight x=0 and y=0 lines
plt.axvline(0, color='black', linewidth=1.5, linestyle='--', label='x = 0')
plt.axhline(0, color='black', linewidth=1.5, linestyle='--', label='y = 0')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = sin(x)')
plt.grid(True)
plt.legend()

plt.show()
```
