### Series In Pandas 
- 1D Labeled Array
- It can hold data of any type like as integers, floats, strings, Python objects.
- components
   - index  (label for each value)
   - values (actual data)


```python
s = pd.Series([23,34,45,78,88])
print(s)
print(type(s))

## indexing 
print(s[0])
print(s[2])
print(s.index)
```

```python
## custom indexing
series = pd.Series([23,44,57,89],index=["A","B","C","D"])
print(series["A"])
print(series["C"])
```

```python
## Vectorized operation
s1 = pd.Series([1,2,3,4])
s2 = pd.Series([7,8,9,10])
print(s1+s2)
```

```python
## Mutable Values but immutable size
s = pd.Series([1,2,3,4,57,8,9])
s[2] = 120
print(s)

changed_series = s.drop(1)
print(changed_series)
```

- In iloc when we slice, the ending index is exclusive.
