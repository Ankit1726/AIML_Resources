```python
import pandas as pd
```

```python
## Question 1 Problem
data = pd.read_csv("iris.csv")
data.head(10)
```

```python
## shape & data type
data.shape
```

```python
data.describe()
```

```python
## Filter Rows
filter_df = data[(data['petal_length']> 4.5) & (data['species'] == "virginica")]
filter_df
```

```python
group_stats = data.groupby('species').agg(
    avg_sepal_length = ('sepal_length', 'mean'),
    max_petal_width  = ('petal_width', 'max'),
    std_sepal_width  = ('sepal_width', 'std')
)

group_stats
```

```python
data['petal_ratio'] = data['petal_length'] / data['petal_width']
```

```python
data.groupby('species')['petal_ratio'].mean()
```
