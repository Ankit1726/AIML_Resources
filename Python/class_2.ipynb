```python
age = int(input('enter your age'))
if age >= 18:
    print('you can derive..!')
else:
    print('you can not derive..!')
```

```python
## Traffic Light
col = input('enter color')
if col == 'red':
    print('Stop..!')
elif col == 'yellow':
    print('Look..!')
elif col == 'green':
    print('Go..!')
else:
    print('wrong color')
```

```python
## Ternary Statement:optimised of if-else (Imp)
age = int(input('enter your age'))
status = "adult" if age>=18 else "not adult"
print(status)
```

```python
## match case statement:aletrnative of longchain if..elif..else statements.

food = input('enter food item')
match food:
    case 'samosa':
        print("5$")
    case 'pizza':
        print('27$')
    case 'burger':
        print('7$')
    case _:
        print('unavailable food..!')
    
```

#### Loops
- It execute block of code multiple time without explicit write code again and again.
   - for: loop-iterate over a sequence (like string or range)
   - while: loop-repeat while logical condition become True. 

```python
i = 1
while i<=7:
    print(i,end=' ')
    i+=1
```

- Loop Control statement
  - break: It stops the loop immediately
  - continue: It skips the current iteration and moves to next one.

```python
i = 1
while i <=10:
    if(i%5==0):
        break
    print(i,end=" ")
    i+=1
```

```python
## Skip multiple of 3
i = 0
while (i<10):
    i+=1
    if(i%3==0):
        continue
    print(i,end=" ")
```

```python
for  i in range(10):
    print(i,end=" ")
```

- in keyword: membership operator in python used in loop over sequence to check presence of an items.

```python
word = "Data Science Engineer"
for ch in word:
    print(ch,end=" ")
```

```python
## count numbers of i in given word
word = 'Artificial Inteligence'
count = 0
for ch in word:
    if ch == 'i':
        count+=1
print(f"i occur {count} times")
```

```python
## Nested Loop
for i in range(1,3):
    for j in range(1,3):
        print(f"({i},{j})")      
```

- range( ): It is used to generate a sequence of numbers having (start,stop,step).

```python
for i in range(1,10,2):
    print(i,end=" ")
```

```python
## Print Table of user input number (using while).
num = int(input('enter your number'))
i = 1
while i < 11:
    print(f"{num} X {i} = {num*i}")
    i+=1
```

```python
## Print odd number from 1-10 (using while)
i = 1
while i<=10:
    i+=1
    if(i%2==0):
        print(i,end=" ")
        i+=1
```

```python
## count vowels in words
word = str(input('enter your words')).lower()
count = 0
for ch in word:
    if(ch in 'aeiou'):
        count+=1
print(f"{word} of total count of vowels: {count}")
```

```python
## Sum of first n natural numbers
num = int(input("enter your number: "))
total = 0

for i in range(1, num + 1):
    total += i

print(f"{num} of natural sum : {total}")
```

#### Functions
- These are block of reusable code that perform specific task (using def keyword).

```python
# function definition
def greet():
    print('Good Morning..!')
greet() # function cal
```

```python
## functions with parameters
def sum(a,b):
    print(a+b)
    
sum(23,21)
```

```python
## return keyword
def avg(a,b,c):
    return (a+b+c)/3

res = avg(21,32,43)
print(f"average of 3 number :{res}")
```

```python
## fnx with default parameter
def sum(a,b=10):
    return a+b

ans = sum(32)
print(f'sum of a & b is {ans}')
```

- Lambda Function
   - One-liner fnx that are used to perform simple tasks (using lambda).

```python
## Square of number
num = int(input('enter your number'))
sq = lambda num:num*num
print(f'square of {num} is {sq(num)}')
```

```python
## function to compute factorial of user input number
def factorial(n):
    fact = 1
    for i in range(1,n+1):
         fact*=i
    return fact

n = int(input('enter your number'))
res = factorial(n)
print(f'factorial of {n} : {res}')
```

```python
## Gwt Largest of 3 numbers
def get_largest(a,b,c):
    if(a>b and a>c):
        return a
    elif b>c:
        return b
    else:
        return c
    
print(get_largest(21,34,55))
```
