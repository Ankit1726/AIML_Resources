- Iterator
  - Iterators are advanced Python concepts that allow for efficient looping and memory management.
  - Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.

```python
lst = [1,2,3,4,5]
for i in lst:
     print(i,end=" ")
```

```python
print(type(lst))
```

```python
## Iterator
iterator = iter(lst)
print(iterator)
print(type(iterator))
```

```python
iterator = iter(lst) 
try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator")
```

```python
## String Iterator 
st = "Hello Good Morning..!"
str_iter = iter(st)

print(next(str_iter)) 
print(next(str_iter))
print(next(str_iter))
```

- Generator 
   - Generators are a simpler way to create iterators.
   -  They use the **yield keyword** to produce a series of values lazily, which means they generate values on the fly and **do not store them in memory**.

```python
def sq(n):
     for i in range(5):
          yield i**2
```

```python
sq(3)
```

```python
for i in sq(5):
     print(i,end=" ")
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
gen
```

- Practical Example: 
     - Reading Large Files Generators are particularly useful for reading large files.
     -  Because they allow you to process **one line** at a time without loading the entire file into memory.

```python
def read_file(file_path):
     with open(file_path,"r") as f:
          for line in f:
               yield line
     
file_path = 'large.txt'
for l in read_file(file_path):
     print(l.strip())
```

- Decorators
   - Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method.
   - They are commonly used to **add functionality** to functions or methods **without** modifying their actual code

- Decorators Methods
   -  function copy
   -  closures
   -  decorators

```python
## function copy
def wel():
     return "Welcome to Data Science World..!"

wel()
```

```python
w = wel
print(w())
del wel
print(w())
```

```python
## closure fnx : The inner function keeps access to the outer function’s variables.

def main(msg):
     def wel():
          print("Welcome to Data Science World..!")
          print(msg)
          print("Hello Senior AI Engineer..!")
     return wel()
```

```python
main("Hello Everyone..!")
```

```python
def outer(x):
    def inner(y):
        return x + y
    return inner

add_five = outer(5)
print(add_five(3))  
```

```python
## Decorator : A decorator is a function that adds extra behavior to another function without changing that function’s code.

def my_deco(func):
     def wrapper():
          print("Before Calling..!")
          func()
          print("After calling..!")
     return wrapper

@my_deco
def greet():
     print("Hello Senior AI Engineer..!")

greet()
```

```python
## Decorator With Argument
def smart_div(fnx):
     def wrapper(a,b):
          if b == 0:
               print("Can not divide by zero")
               return
          return fnx(a,b)
     return wrapper

@smart_div
def divide(a,b):
     return a/b

print(divide(10, 2))
print(divide(10, 0))
```

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Execution time: {end - start:.4f}s")
        return result
    return wrapper
```

```python
@timer
def slow_function():
    time.sleep(2)
    print("Function executed")

slow_function()
```
