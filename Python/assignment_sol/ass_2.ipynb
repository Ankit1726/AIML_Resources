```python
## Ques_1
salary = float(input("enter your salary"))
if salary < 30000:
     tax = salary * 0.05
elif salary >=30000 and salary < 70000:
     tax = salary * 0.15
elif salary >=70000:
     tax = salary * 0.25
else:
     print('invalid salary')
print(f"Final Tax: {tax}")
```

```python
## Ques_2
def printEven(a,b):
     for i in range(a,b+1):
          if(i%2==0):
               print(i,end=" ")

a = int(input('enter your number '))      
b = int(input('enter your number '))    
print(f"All Even Numbers B/W {a} & {b}")
printEven(a,b)
```

```python
## Ques_3
def print_digits(n):
    while n > 0:
        digit = n % 10
        print(digit,end=" ")
        n = n // 10

n = int(input('enter your digit'))
print_digits(n)
```

```python
## Ques_4
def countNum(n):
     count = 0
     while n>0:
          count+=1
          n = n//10
     return f'Total Count : {count}'

n = int(input("enter your digit"))
ans = countNum(n)   
print(f"Total Count of Entered {n} number is {ans}")
```

```python
## Ques_5
def sum_digit(n):
    total = 0
    while n > 0:
        digit = n % 10
        total += digit 
        n = n // 10   # remove last digit
    return total

n = int(input("Enter your number: "))
print(f"Sum of Digits: {sum_digit(n)}")
```

```python
## Ques_6
def printNum(n):
     for i in range(1,n+1):
          if (i%3==0 and i%5==0):
               print(i,end=" ")
print("Number Divisible By 3 and 5")
printNum(100)
```

```python
## Ques_7
while True:
    user_input = input("Enter a number (or type Quit to exit): ")

    if user_input.lower() == "quit":
        print("Program terminated.")
        break

    num = float(user_input)

    if num > 0:
        print("Positive number")
    elif num < 0:
        print("Negative number")
    else:
        print("Zero")
```

```python
## Ques_8

def calculator(a,b,operation):
     match operation:
          case "+":
               print(f"Addition:{a+b}")
          case "-":
               print(f"Substraction:{a-b}")
          case "*":
               print(f"Multiplication:{a*b}")
          case "/":
               print(f"Division:{a/b}")
          
a = int(input("enter your number "))
b = int(input("enter your number "))
operation = input("enter your operation to be performed")
calculator(a,b,operation)
```

```python
## Ques_9
def isPrime(n):
     for i in range(2,n+1):
          if(i%2==0):
               return True
          return False
     

n = int(input("enter your number"))
isPrime(n)
```

```python
## Ques_10

def GuessGame():
    secret_num = 67
    while True:
        num = int(input("Enter your guess: "))

        if num > secret_num:
            print("Too High")
        elif num < secret_num:
            print("Too Low")
        else:
            print("Correct")
            break

GuessGame()
```
