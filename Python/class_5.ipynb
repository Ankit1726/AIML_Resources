- File I/O 
   - File Input/Output means reading data from files & writing data to files using Python.

- Opening File: open()
   - common mode
      - r
      - w
      - a
      - r+
      - w+
      - a+
      - b

```python
## Reading data from file [read()]
f = open("data.txt","r")
content = f.read()
print(content)
f.close()
```

```python
## read a line at time [readline()]
f = open("data.txt","r")
l1 = f.readline()
l2 = f.readline()
print(l1)
print(l2)
f.close()
```

```python
## read all lines into list [readlines()]
f = open("data.txt","r")
lines = f.readlines()
print(lines)
f.close()
```

```python
## Writing to a file [write()]
f = open("sample.txt","w")
w = f.write("Hello Senior AI Engineer..!")
print(w)
f.close()
```

```python
## writing multiple lines at once [writelines()]
f = open("sample.txt","w")
res = f.writelines(["Hello Good Morning \n","I am Senior AI Engineer at Google \n","How are you all developers..!"])
f.close()
```

- To Following Best Practices  Usage with open() it used as a context manager that automatically closes file.

```python
with open("sample.txt","r") as f:
     content = f.read()
     print(content)
```

```python
## Deleting a file used [remove()]
import os
res = os.remove("sample.txt")
print(res)
```

- Exception Handling
  - An exception is an error that occurs while a program is running. 
  - If not handled the program crashes.
 - Error Encounter:
    - Dividing by zero : ZeroDivisionError
    - Using an undefined variable : NameError
    - Opening a missing file : FileNotFoundError
    - Wrong data type : TypeError

```python
## try & except
try:
     x = 10/0
except:
     print("errot occur..!")
```

```python
try:
     x = 10/0
     print(x)
except ZeroDivisionError:
     print("You can't divide by zero..!")
```

```python
## try & except with else it executes only when no exception happens.

try:
     x = int(input("enter your number"))
except ValueError:
     print("Invalid Input")
else:
     print("You Entered:",x)
```

```python
## Finally block: It always executes,wheather an exception occur or not.
try:
     f = open("data.txt")
     print(f.read())
except FileNotFoundError:
     print("Fiel not found...!")
finally:
     print("Execution Completed..!")
```

- List Comprehensions
     - List Comprehension is a short and elegant way to create lists in Python.
     - [ expression for item in iterable ].

```python
## create a list of numbers 1 to 5
nums = [x for x in range(1,6)]
print(nums)
```

```python
## Even numbers from 1 to 10
even = [x for x in range(1,11) if x%2==0]
print(even)
```

```python
## Label numbers as “Even” or “Odd”
labels = ["even" if x%2==0 else "odd" for x in range(1,6)]
print(labels)
```

- JSON Module
  - JSON (JavaScript Object Notation) is a lightweight data format used to exchange data between programs, APIs, websites, etc.
  - JSON Similar to Python Dictionaries.

```python
import json
```

```python
## Converting Python Data to JSON Data use (dumps())
data = {
"name": "John",
 "age": 25,
 "marks": [85, 90, 92]
}
json_data = json.dumps(data)
print(json_data)
print(type(json_data))
```

```python
## Converting JSON Data To Python Used (loads())
json_data = '{"name": "John", "age": 25}'
python_obj = json.loads(json_data)
print(python_obj)
print(type(python_obj))
```

```python
## Reading JSON Data From File Used (json.load())
import json 
with open("record.json","r") as f:
     data = json.load(f)

print(data)
```

```python
##Writing JOSN Data To File Used (json.dump)
import json
data = {
      "student_id": 101,
      "name": "Ankit Gupta",
      "age": 21,
      "gender": "Male",
      "course": "BTech",
      "math_score": 85,
      "science_score": 78,
      "english_score": 88,
      "attendance_percentage": 92,
      "passed": True
    }

with open("student.json","w") as f:
     record = json.dump(data,f,indent=3)

print(record)
```
