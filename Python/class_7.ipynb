### SQL & SQLite
- **SQL** (Structured Query Language) is a standard language for managing and manipulating relational databases.
-  **SQLite** is a self-contained, serverless, and zero-configuration database engine that is widely used for embedded database systems.

```python
import sqlite3
```

```python
## Connect to an SQLite database
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
## Create a Table
cursor.execute('''
Create Table If Not Exists employees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department text
    )
''')

## commit the changes
connection.commit()
```

```python
cursor.execute('''
Select * from employees 
''')
```

```python
## Insert the data in sqlite table
cursor.execute('''
Insert Into employees(name,age,department)
               values
               ('Krish',32,'Data Scientist'),
               ('Ankit',21,'AI Engineer'),
               ('Kush',22,'ML Engineer');
''')
```

```python
cursor.execute('''
INSERT INTO employees (name, age, department)
VALUES ('Charlie', 35, 'Finance')
''')
```

```python
## commi the changes
connection.commit()
```

```python
## Query the data from table
cursor.execute('Select * from employees')
rows = cursor.fetchall()

##print Queried data
for r in rows:
     print(r)
```

```python
## Update the data in the table
cursor.execute('''
UPDATE employees
Set age=34
where name="Krish"
''')
```

```python
## Query the data from the table
cursor.execute('Select * from employees')
rows=cursor.fetchall()

## print data 
for r in rows:
     print(r)
```

```python
## Delete the data from the table
cursor.execute('''
Delete from employees
               where name ='Bob'
''')

connection.commit()
```

```python
## Query the data from the table
cursor.execute('Select * from employees')
rows=cursor.fetchall()

## print data 
for r in rows:
     print(r)
```

```python
## Worling With Sales Data
# connect to sqlite database
connection = sqlite3.connect("Sales_Data.db")
cursor = connection.cursor() # cursor help to connect with sql
```

```python
## create table for sales data
cursor.execute('''
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    product TEXT NOT NULL,
    sales INTEGER,
    region TEXT
)
''')
```

```python
# Insert data into the sales table
sales_data = [
    ('2023-01-01', 'Product1', 100, 'North'),
    ('2023-01-02', 'Product2', 200, 'South'),
    ('2023-01-03', 'Product1', 150, 'East'),
    ('2023-01-04', 'Product3', 250, 'West'),
    ('2023-01-05', 'Product2', 300, 'North')
]
```

```python
cursor.executemany('''
Insert into sales(date,product,sales,region)
                   values(?,?,?,?)
''',sales_data)

connection.commit()
```

```python
# Query data from the sales table
cursor.execute('SELECT * FROM sales')
rows = cursor.fetchall()

# Print the queried data
for row in rows:
    print(row)
```

```python
## close the connection
connection.close()
```
