- Constructor in OOP
  - It is a special method used to initialize newly created objects.
  - we used __init__(self,...).
  - self: it refers current object or instance.

```python
class Constructor:
    def __init__(self): # Default Constructor
        print('constructor was called..!')
    
call = Constructor()
print(call)
```

- Types of Constructor
   - Default: no parameters except (self).
   - Parameterized: take parameters to initialize value uniquely for each objects.

```python
class Student:
    def __init__(self,name):  # Parameterised Constructor
        self.name = name
    
s1 = Student("Ankit")
s2 = Student("Kush")

print(s1.name)
print(s2.name)
```

- Attributes: These are variables.
- Methods: These are functions defined inside a class.
 
- Types of Attributes
    - class attributes
        - belongs to **class itself**.
        - shared among all objects.
        - defined **outside** any method in the class.
    - instance attributes
        - Belong individually to **each object**.
        - defined inside **__init__** method using **self**.
        - each object gets its own copy.

```python
## class attribute
class Student:
    college = "Indus Uiversity" # class attribute

s = Student()
print(s.college)
print(Student.college)
# class attribute can also be accessed with class name 
```

```python
## Instance Attribute
class Student:
    def __init__(self,name,gpa): # istance attribute
        self.name = name
        self.gpa = gpa

s = Student("Ankit",8.0)
print(s.name,s.gpa)
```

- Methods in OOP
   - These are functions defined inside a class,
   representing the **behaviour** or **actions** of an object.
   - Tyes of Methods
     - Instance Methods
     - Class Methods
     - Static Methods

```python
## instance method
class data:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

    def display(self): # instance
        print(f"Name:{self.name} Marks:{self.marks}")
d1 = data("Ankit",97)
d2 = data("kush",94)

d1.display()
d2.display()
```

```python
## Class Method
class student:
    school_name = "American Times"

    @classmethod # decorator
    def change_school(cls,new_name):
        cls.school_name = new_name

print(student.school_name)
student.change_school("IIT Delhi")
print(student.school_name)
```

```python
## static methods
class Math:
    @staticmethod
    def add(a,b):
        return a+b

res = Math.add(10,20)
print(res)
```

- Encapsulation
   - data(variable) + methods(fnx) = single unit(class),along with control access of data.
- To Implement Encapsulation
    - use access modifier named as:
      - Public 
      - Protected:( _ )
      - Private:( __ )

```python
## public members
class Student:
    def __init__(self,name):
        self.name = name # public variable
    
s = Student("Ankit")
print(s.name)
```

```python
## protected members
class Person:
    def __init__(self):
        self._age = 21  # protected var

p = Person()
print(p._age) # not good practice.
```

```python
## private members
class Bank:
    def __init__(self,bal):
        self.__bal = bal # private var

B = Bank(54007)
# print(b. balance) Error
print(B._Bank__bal)
```

```python
## To access private variable we can use getter & setter.

class Employee:
    def __init__(self, salary):
        self.__salary = salary  # private

    def get_sal(self):
        return self.__salary  # getter

    def set_sal(self,new_salary): # setter
        self.__salary = new_salary
    
e = Employee(79090)
print(e.get_sal())
e.set_sal(9900000)  # set salary
print(e.get_sal())  # get salary
```

- Inheritance
   - child class inherit properties & behaviour of parent class or another class.
   - It enables us: 
       - code reusability
       - Extensibility
   
   - Terms to be used:
      - Parent | Base | Superclass.
      - Child | Derived | Subclass.

```python
## parent class
class Emp:
    start_time = "9AM"
    end_time = "5PM"

## child class
class Teacher(Emp):
    def __init__(self,subject):
        self.subject = subject

t1 = Teacher("Data Science With GenAI")
print(t1.subject)
print(f" Start Time: {t1.start_time}")
print(f" End Time: {t1.end_time}")
```

- Types of Inheritance
  - Single 
  - Multi-level
  - Multiple

```python
## Single Inheritance (parent --> child)
class Parent:
    def display(self):
        print('Parent class..!')

class Child(Parent):
    pass

c = Child()
c.display()
```

```python
## Multi-level Inheritance (grandparent --> parent --> child)

class Employee:
    st_time = '9AM'
    ed_time = '5PM'

class AdminStaff(Employee):
    def __init__(self,role):
        self.role = role

class Accountant(AdminStaff):
    def __init__(self,salary, role):
        super().__init__(role)
        self.salary = salary

acc = Accountant(5000000,"Data Scientist")
print(acc.salary)
print(acc.role)
print(acc.st_time)
print(acc.ed_time)
```

- **super()** keyword : Used to call parent class method from child class.

```python
## Multiple Inheritance
class Teacher:
    def __init__(self,salary):
        self.salary = salary

class Student():
    def __init__(self,gpa):
        self.gpa = gpa

class TA(Teacher,Student):
    def __init__(self,name,salary,gpa):
        super().__init__(salary) # parent call
        Student.__init__(self,gpa)
        self.name = name
    
res = TA("Ankit",340000,7.5)
print(res.name)
print(res.salary)
print(res.gpa)
```

- Abstraction
    - Hiding unnecessary implementation details & showing only essential features to the user.
    - It can not instansiated.
    - It contains abstract methods.
    - It acts as a blueprint for child classes.

```python
from abc import ABC,abstractmethod
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

class Lion(Animal):
    def make_sound(self):
        print("Roar..!")

class Cow(Animal):
    def make_sound(self):
        print('Moo..!')

cow = Cow()
cow.make_sound()
lion = Lion()
lion.make_sound()
```

- Polymorphism
   - Having Multiple Forms.
   - ability of a single function or object to behave differently on the context.
- Polymorphism Type
   - Function Overriding
      - Child method takes more precedance over parent method while method have same name but differ in functionalities.
   - Duck Typing
       - If it looks like a duck and quacks like a duck, Python treats it as a duck. 
       - It means that python does not care about class of an object it only cares wheather the object has required methods or behaviour.

```python
## Duck Typing  
class Dog:
    def speak(self):
        print("Bark..!")
    
class Cat:
    def speak(self):
        print('meow..!')
    
class Robot:
    def speak(self):
        print("Beep Boop..!")

def make_speak(entity):
    entity.speak() # not care about type

d = Dog()
c = Cat()
r = Robot()

for e in [d,c,r]:
    make_speak(e)
```
