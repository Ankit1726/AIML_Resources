- String
   - It is a sequence of character enclosed in quotes.
   - These are immutable datatype.

```python
str1 = "Hello Good Morning..!"
print(str1)
print(type(str1))
```

- len() : It is used to find out length of string.

```python
word = "Prime AI"
print(len(word))
```

- Concatenation :adding two string '+' using this make single string.

```python
str1 = "Hello"
str2 = "World"
word = str1+" "+str2
print(word)
```

```python
## Loop on String
s = 'Python'
for ch in s:
    print(ch,end=" ")
```

- Indexing in Strings
   - Index in a sequence (like strings) represents the position value pf individual element and follows zero-based indexing.

```python
s = "Data Scientist"
print(s[0])
print(s[-1])
```

- Slicing in strings
  - It access multiple elements at once.
  - having string(start:stop:step).

```python
str1 = "Data Science"
print(str1[:])
print(str1[::-1])
print(str1[::2])
```

- String Formating: 
   - It is the proccess of creating dynamic strings by inserting values expressions into predefined string template.
     - format()
     - f-strings()

```python
## format()
language = 'Python'
domain = 'Data Science'
txt = "I am Ankit and learning {} & belongs to {}".format(language,domain)
print(txt)  
# It follows order of input sequence.
```

```python
## f-string()
a = 34
b = 21
print(f"sum of {a} & {b} : {a+b}")
```

- List
   - A list is a collection of items in Python
   - Items in a list are ordered,mutable & allow duplicates.
   - It can hold any(hetrogeneous) datatype.
   - Using [ ] for creating list

```python
my_lst = [1,2,3,4,5]
print(my_lst)
print(type(my_lst))
```

```python
## List Indexing 
lst = ['apple','cherry','grapes','banana']
print(lst[0])
print(lst[2])
```

```python
## modify elements
lst = [1,2,3,4,5]
lst[0] = 10
print(lst)
```

- list slicing
   - using syntax list[start:end:step].

```python
## list slicing
num = [0,1,2,3,4,5,7,8,9]
print(num[2:])
print(num[::-1])
print(num[:])
print(num[-2::])
```

- List Methods
   - len()
   - append(element)
   - insert(element,idx)
   - sort()
   - reverse()

```python
number = [2,3,4,5,8]
print(len(number))

number.append(8)
print(number)

number.insert(2,9)
print(number)

number.sort()
print(number)

number.reverse()
print(number)
```

```python
## Loops on Lists
nums = [10,20,30,40,50]
for i in nums:
    print(i,end=" ")
```

```python
## Linear Search
def linear(n,target):
    idx = 0
    for num in n:
        if (num == target):
            print(f"{target} found at index:{idx}")
            break
        idx+=1
      
n = [10,20,30,43,45,57,89]
linear(n,30)
```

- Tuple
  - It is an ordered.
  - Immutable collection of items.
  - Allow duplicates.
  - Hetrogeneous(multiple datatype).

```python
## Indexing & Slicing
tup = (10,20,30,45,78)
print(tup[0])
print(tup[:])
print(tup[::-1])
print(tup[:4:])
```

```python
## Loops on Tuple
tup = (12,23,34,45,57)
for val in tup:
    print(val,end=" ")
```

```python
## sum of all element in tuple
t = (5,10,15,20)
sum = 0
for i in t:
    sum+=i
print('Total Sum :',sum)
```

- Tuple Methods
  - index(val) : returns index of first occur.
  - count(val) : returns total count of occur.

```python
tup = (1,2,3,4,5,7,8,9,3,2,1,4,3)
print(tup.index(4))
print(tup.count(4))
```

- Dictionary
   - It is an unordered,mutable collection of key-val pairs.
   - Dictionary keys must be unique & immutable.
   - Dictionary is mutable datatype.
   - Hetrogenous in nature.
   - Dictionary not follows any order sequence.

```python
my_dict = {
    "name":"Ankit",
    "age":21,
    "city":"Noida"
}
print(my_dict)
print(my_dict.keys())
print(my_dict.values())
```

```python
## Accessing Values
student = {"name": "Bob", "age": 20}
print(student["name"])
```

- Dictionary Methods
  - keys()
  - values()
  - items()
  - get(key)
  - update(new_item)

```python
data = {
    "name":"Ankit",
    "subjects": ["math", "science", "physics"],
    "cgpa":8.0
}
print(data.keys())
print(data.values())
print(data.items())
print(data.get("subjects"))
```

```python
new_item = {"city":"Noida"}
print(data.update(new_item))
print(data)
```

```python
## Loop on Dictionary
d = {
    'name':'Ankit',
    'subjects':['maths','science','physics'],
    'cgpa':9.0
}

for key,val in d.items():
    print(key,val)
```
