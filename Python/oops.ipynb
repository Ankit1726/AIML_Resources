```python
## Parameterized Constructor
class Student:
    def __init__(self,name):
        self.name = name
    
stu = Student("Ankit")
print(stu.name)
```

```python
## Attributes
class Student:
    college = "NIT Kanpur" # class Attribute
    def __init__(self,name,gpa): # instance attribute
        self.name = name
        self.gpa = gpa

stu = Student("Rahul",8.0)
print(f"{stu.name} studying in {stu.college} has {stu.gpa} cgpa")
```

```python
## class attribute can accessed with class name
print(f"{stu.name} studying in {Student.college} has {stu.gpa} cgpa")
```

```python
## Methods
class Laptop:
    storage_type = "SSD"
    def __init__(self,RAM):
        self.RAM = RAM
    
    def get_info(self): # instance method
        print(f"Laptop has {self.RAM} RAM & Storage Type is {self.storage_type}")

    @classmethod  
    def get_storage_type(cls):
        print(f"storage type is {cls.storage_type}")
    @staticmethod
    def cal_discount(price,discount):
        final_price = price - (price * discount / 100)
        print(f"Final Price = {final_price}")

L1 = Laptop("12 GB")
L1.get_info()
Laptop.get_storage_type()
L1.cal_discount(70_000,10)
```

```python
## Encapsulation
class BankAccount:
    def __init__(self,name,bal):
        self.name = name  # public
        self.__bal = bal  # private

    def get_bal(self):
        return self.__bal  # getter

    def set_bal(self,new_bal):
        self.__bal = new_bal # setter

acc = BankAccount("Ankit",1_00_0000)
print(acc.get_bal())
acc.set_bal(70_000_0000)
print(acc.get_bal())
```

```python
## Inheritance - single Inheritance
class Employee: # parent class
    st_time = "9am"
    end_time = "5am"

class Teacher(Employee): # child class
    def __init__(self,subject):
        self.subject = subject

T = Teacher("Data Science")
print(T.subject,T.st_time,T.end_time)
```

```python
# Multiple Inheritance

class Teacher:             
    def __init__(self, salary):
        self.salary = salary

class Student():     
    def __init__(self, gpa):
        self.gpa = gpa

class TA(Teacher, Student):    
    def __init__(self, name, salary, gpa):
        super().__init__(salary)
        Student.__init__(self, gpa)
        self.name = name

ta = TA("Rahul", 50_000, 7.5)

print(ta.name, ta.salary, ta.gpa)
```

```python
## Abstraction
from abc import ABC,abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound():
        pass

class Lion(Animal):
    def make_sound(self):
        print("Roar..!")

class Cow(Animal):
    def make_sound(self):
        print("Moo..!")

lion = Lion()
lion.make_sound()

cow = Cow()
cow.make_sound()
```

```python
## Polymorphism

## Opeator Overlaoding
print(1+3)
print("1"+"3")

## Function Overriding
class Animal:
    def sound(self):
        print("some generic sound..!")

class Dog(Animal):
    def sound(self):
        print("Dog is Barking..!")

ani = Animal()
ani.sound()
dog = Dog()
dog.sound()
```

```python
## Duck Typing
class Dog:
    def speak(self):
        print("Barking..!")

class Cat:
    def speak(self):
        print("Meow..!")

class Robot:
    def speak(self):
        print("Beep Bop..!")

def make_it_speak(entity):
    entity.speak()  # don't care about type

d = Dog()
c = Cat()
r = Robot()

for e  in [d,c,r]:
    make_it_speak(e)
```
